## Loops

```python
# Using range

range(3)
```

```python
# Using for loop

date = [1982, 1980, 1973]
N = len(date)

for i in range(N):
    print(date[i])
    
```

```python
# Example of for loop through list

for year in date:
    print(year)
```

```python
# Example of for loop 

for i in range(10, 15):
    print(i)
```

```python
# Using for loop to change an element in a list

squares = ['red', 'yellow', 'green', 'purple', 'blue']

for i in range(0,5):
    print('Before change ', i , 'is', squares[i])
    squares[i] = 'white'
    print('After change ', i , 'is', squares[i])

squares
```

```python
# Loop through the list and iterate on both index and element value

for n, square in enumerate(squares):
    print(n , square)
```

```python
# while loop example

dates = [1982, 1980, 1973, 2000]

i =0
year = dates[0]

while year != 2000:
    print(year)
    i = i+1
    year = dates[i]
    
    
print('It took ', i, ' repetitions to get out of the loop')
```
